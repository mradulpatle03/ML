Convert to datetime

```python
df['ScheduledDay'] = pd.to_datetime(df['ScheduledDay'])
df['AppointmentDay'] = pd.to_datetime(df['AppointmentDay'])
```

```python
(df['AppointmentDay'] < df['ScheduledDay']).sum()
```

```python
df['ScheduledDate'] = df['ScheduledDay'].dt.date
df['AppointmentDate'] = df['AppointmentDay'].dt.date

(df['AppointmentDate'] < df['ScheduledDate']).sum()
```

Although initial checks showed appointments occurring before scheduling, this was due to inconsistent timestamp. AppointmentDay is recorded at midnight, while ScheduledDay includes time. After normalizing both to date-only values, no logical inconsistencies remained.

```python
df['waiting_days'] = (
    df['AppointmentDate'] - df['ScheduledDate']
).apply(lambda x: x.days)
```
